```python
import os
from ultralytics import YOLO
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Path ke gambar
image_path = "./testing_data/cat5.jpg"

# Load the image
image = cv2.imread(image_path)

# model_path = "./runs/detect/train/weights/last.pt"
model_path = "./hasil-collab/detect/train2/weights/best.pt"
model = YOLO(model_path)

threshold = 0.5

# Run the model on the image
results = model(image)[0]

# Plot the image with predictions
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.axis('off')

# Menambahkan kotak dan label untuk setiap prediksi
for result in results.boxes.data.tolist():
    x1, y1, x2, y2, score, class_id = result

    if score > threshold:
        plt.gca().add_patch(plt.Rectangle((x1, y1), x2 - x1, y2 - y1, fill=False, edgecolor='red', linewidth=2))
        plt.text(x1, y1 - 5, results.names[int(class_id)].upper() + ': {:.2f}'.format(score),
                 bbox=dict(facecolor='red', alpha=0.5), fontsize=8, color='white')

# Menampilkan plot
plt.show()
```
